```python
violations = frame_query("SELECT * from violations ", conn)
violations = violations.drop(["restaurant_id","created_date"],axis=1)
Xtrain = violations.drop(['violation_id'],axis=1).values
Ytrain = violations.drop(['minor_violation_score','major_violation_score','serve_violation_score'],axis=1).values
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.cross_validation import train_test_split
x_train,x_test,y_train,y_test = train_test_split(Xtrain,Ytrain.ravel())
rf = RandomForestRegressor()
rf.fit(x_train,y_train)
prediction= rf.predict(x_test)
print(rf.feature_importances_)
```
